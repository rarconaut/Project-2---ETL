```python
# Dependencies
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Name of the CSV file
file = 'Resources/ds442.csv'
```

```python
# The correct encoding must be used to read the CSV in pandas
df = pd.read_csv(file, encoding="ISO-8859-1")
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['ACT_DATE'] =  pd.to_datetime(df['ACT_DATE'], infer_datetime_format=True) 
```

```python
df['SECTION'] = df['SECTION'].astype(str)
df["TRANSNAME"] = df["TRANSNAME"] + " - Section: " + df["SECTION"]
```

```python
clean_df = df[["SNAME","CNAME","ACT_DATE","TRACK_SIGN","TOT_SIGHT","VEG_CAT","TRANSNAME","SITE","LAT_MID","LONG_MID"]]
```

```python
clean_df = clean_df.rename(columns={"SNAME": "species_scientific_name", "CNAME": "species_common_name","ACT_DATE":"observation_date","TRACK_SIGN":"track_type",
                   "TOT_SIGHT":"total_sightings","VEG_CAT":"vegetation_category","TRANSNAME":"location","SITE":"reserve_name",
                   "LAT_MID":"lat","LONG_MID":"long"})
clean_df.head()
```

```python
username = input("Enter Username: ")
password = input("Enter Password: ")
connection_string = username + ":" + password + "@localhost:5432/project2_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
clean_df.to_sql(name='species_scientific_name', con=engine, if_exists='append', index=True)
```
