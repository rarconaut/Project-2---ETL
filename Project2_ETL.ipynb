```python
import pandas as pd
from sqlalchemy import create_engine
```

### Extract CSVs into DataFrames

```python
city_file = "Resources/city_data.csv"
city_df = pd.read_csv(city_file, low_memory=False)
pd.set_option('display.max_columns', 50)
city_df.head(150)
```

```python
mammal_file = "Resources/mammal_data.csv"
mammal_df = pd.read_csv(mammal_file)
pd.set_option('display.max_rows', 150)
mammal_df.head(150)
```

### Transform city DataFrame

```python
# Create a filtered dataframe from specific columns
city_cols = ['project_id', 'project_status', 'date_project_create', 'date_project_complete', 'project_title', 
             'project_scope', 'job_id', 'address_job', 'job_bc_code', 'job_bc_code_description', 'lat_job', 'lng_job', 
             'approval_type', 'approval_valuation', 'council_district', 'zip']
city_transformed = city_df[city_cols].copy()

# # Rename the column headers
# city_transformed = city_transformed.rename(columns={"License Serial Number": "id",
#                                                           "Premises Name": "premise_name",
#                                                           "County ID Code": "county_id"})

# Clean the data by dropping duplicates and setting the index
# city_transformed.drop_duplicates("id", inplace=True)
city_transformed.set_index("project_id", inplace=True)

city_transformed.head()
```

```python
major_construction = city_transformed[city_transformed.approval_valuation > 10000, city_transformed.job_bc_code = 'NaN']
major_construction
```

### Transform mammal DataFrame

```python
mammal_cols = ['SNAME', 'CNAME', 'ACT_DATE', 'TRACK_SIGN', 'TOT_SIGHT', 'VEG_CAT', 'TRANSID', 'SECTION', 
               'TRANSNAME', 'SITE', 'LAT_MID', 'LONG_MID']
mammal_cols = mammal_df[mammal_cols].copy()

# Rename the column headers
mammal_transformed = mammal_df.rename(columns={'SNAME': 'species', 
                                                        'CNAME': 'common_name', 
                                                        'ACT_DATE': 'date', 
                                                        'TRACK_SIGN': 'track_sign', 
                                                        'TOT_SIGHT': 'total_sight', 
                                                        'VEG_CAT': 'habitat', 
                                                        'TRANSID': 'transid', 
                                                        'SECTION': 'section', 
                                                        'TRANSNAME': 'transect', 
                                                        'SITE': 'location', 
                                                        'LAT_MID': 'lat', 
                                                        'LONG_MID': 'lng'})

# Set index
mammal_transformed.set_index("transid", inplace=True)

mammal_transformed.head()
```

### Create database connection

```python
connection_string = "postgres:postgres@localhost:5432/project2_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
# Confirm tables
engine.table_names()
```

### Load DataFrames into database

```python
city_transformed.to_sql(name='city_data', con=engine, if_exists='append', index=True)
```

```python
mammal_transformed.to_sql(name='mammal_data', con=engine, if_exists='append', index=True)
```
